```python
import os
import sys
sys.path.append("..")
import torch
import pickle
from tqdm.notebook import tqdm
from src.model.net import SAN
from REMI.midi2event import analyzer, corpus, event
```

```python
model = SAN(num_of_dim= 4, vocab_size= 339, lstm_hidden_dim= 128, embedding_size= 300, r=4)
checkpoint_path = "../exp/PEmo/SAN_REMI_head4/av_remi_gpu1/epoch=96-val_loss=0.8408-val_acc=0.6715.ckpt"
state_dict = torch.load(checkpoint_path, map_location=torch.device("cuda:1"))
new_state_map = {model_key: model_key.split("model.")[1] for model_key in state_dict.get("state_dict").keys()}
new_state_dict = {new_state_map[key]: value for (key, value) in state_dict.get("state_dict").items() if key in new_state_map.keys()}
model.load_state_dict(new_state_dict)
model.eval()
model = model.to("cuda:1")
```

```python
path_data_root = "../REMI/ailab17k_from-scratch_remi"
path_dictionary = os.path.join(path_data_root, 'dictionary_with_key.pkl')
midi_dictionary = pickle.load(open(path_dictionary, "rb"))
event_to_int = midi_dictionary[0]
```

```python
midi_path = "../dataset/sample_data/Sakamoto_MerryChristmasMr_Lawrence.mid"
midi_obj = analyzer(midi_path)
song_data = corpus(midi_obj)
event_sequence = event(song_data)
quantize_midi = [event_to_int[str(i['name'])+"_"+str(i['value'])] for i in event_sequence]
```

```python
len(quantize_midi)
```

```python
bar = []
for idx, i in enumerate(event_sequence):
    if i['name'] ==  "Bar":
        bar.append(idx)
```

```python
len(bar)
```

```python
batch = []
events_list = []
splice_idx = 0
for idx, i in enumerate(bar):
    if idx % 12 == 0:
        if idx == 108:
            pass
        else:
            if splice_idx != 96:
                batch.append(quantize_midi[bar[splice_idx]:bar[splice_idx + 12]])
                events_list.append(event_sequence[bar[splice_idx]:bar[splice_idx + 12]])
                splice_idx = splice_idx + 12
            else:
                batch.append(quantize_midi[bar[splice_idx]:bar[splice_idx + 11]])
                events_list.append(event_sequence[bar[splice_idx]:bar[splice_idx + 11]])
                
    else:
        pass
```

```python
song_level = []
song_att = []
for bar_event in tqdm(batch):
    torch_midi = torch.LongTensor(bar_event).unsqueeze(0)
    prediction = model(torch_midi.to("cuda:1"))
    _, attn_mat = model._get_attention_weight(torch_midi.to("cuda:1"))
    song_att.append(attn_mat.squeeze(0).sum(axis=0).detach().cpu().numpy())
    song_level.append(prediction.squeeze(0).detach().cpu().numpy())
```

```python
import numpy as np
from matplotlib import pyplot as plt
song_predict = np.stack(song_level)
Q1_prob = song_predict[:,0]
Q2_prob = song_predict[:,1]
Q3_prob = song_predict[:,2]
Q4_prob = song_predict[:,3]
segment_bin = np.arange(len(song_level)) + 1
```

```python
plt.figure(figsize=(10, 4))
plt.plot(segment_bin, Q1_prob,marker='.', label="Q1")
plt.plot(segment_bin, Q2_prob,marker='.', label="Q2")
plt.plot(segment_bin, Q3_prob,marker='.', label="Q3")
plt.plot(segment_bin, Q4_prob,marker='.', label="Q4")
plt.legend(loc = 1)
plt.xlabel('bar')
plt.xticks(segment_bin)
plt.ylabel('Prob')
plt.show()
```
